```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(100, noise=0.25,random_state=2)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:,0], X[:,1], c=y)
plt.show()
```

```python
model1 = Sequential()

model1.add(Input(shape=(2,)))
model1.add(Dense(128, activation="relu"))
model1.add(Dense(128, activation='relu'))
model1.add(Dense(1,activation='sigmoid'))

model1.summary()
```

```python
adam = Adam(learning_rate=0.01)

model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history1 = model1.fit(X,y, epochs=2000, validation_split=0.2, verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
model2 = Sequential()

model2.add(Input(shape=(2,)))
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l2(0.9)))
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l2(0.9)))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
adam = Adam(learning_rate=0.01)

model1.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history1 = model1.fit(X,y, epochs=2000, validation_split=0.2, verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model1, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
model1_weight_layer1 = model1.get_weights()[0].reshape(256)
model2_weight_layer1 = model2.get_weights()[0].reshape(256)
```

```python
sns.boxplot(model1_weight_layer1)
```

```python
sns.boxplot(model2_weight_layer1)
```

```python
sns.histplot(model1_weight_layer1,kde=True)
sns.histplot(model2_weight_layer1,kde=True)
```
